```python
def cluster_alasan(text):
    if isinstance(text, str):
        text = text.lower()  # Ubah ke huruf kecil untuk pencarian case-insensitive
        if any(keyword in text for keyword in ["meninggal", "wafat", "kematian", "ahli waris", "dilanjutkan", "tanggung", "dunia", "klaim", "asuransi"]):
            return "Meninggal"
        elif any(keyword in text for keyword in ["asing"]):
            return "Nasabah Asing"
        elif any(keyword in text for keyword in ["npwp", "perusahaan", "badan usaha"]):
            return "Badan Usaha"
        elif any(keyword in text for keyword in ["tutup buku", "ph", "hapus buku", "hapusbuku", "hapus tagih", "dihapus buku", "dihapus tagih", "wo", "dihapusbuku", "nasabah sudah tidak bertransaksi", "macet"]):
            return "Hapus Buku/Macet"
        elif any(keyword in text for keyword in ["cif"]):
            return "Merge CIF"
        elif any(keyword in text for keyword in ["e-ktp", "akta lahir", "elektronik", "nik tidak terdaftar", "ktp tidak ada", "ktp debitur", "dokumen kependudukan", "no.identitas", "perbaikan nik", "penyesuaian nik", "kartu penduduk", "nik debitur", "nama", "tanggal lahir", "nama ibu kandung", "nomor ganda", "tidak aktif", "pengkinian", "pemadanan", "peninjauan", "disduk", "tidak ada nik", "kesalahan penginputan", "data tidak sesuai", "ketidaksesuai", "data yang berbeda", "verifikasi dukcapil", "tidak terdaftar di dukcapil", "anomali nik"]):
            return "Masalah Identitas"
        elif any(keyword in text for keyword in ["pencarian", "meminta", "cari","terbaru", "mencari data", "belum dapat data", "belum ketemu", "menunggu", "belum diperoleh", "belum ditemukan", "belum dapat di temui", "blm dapat di temui", "diupayakan", "masih mencari nasabah", "konfirmasi", "sementara mencari", "kunjungan ke debitur", "nasabah kta", "dicari", "validasi", "mencocokkan", "berkas tidak terbaca", "nasabah non aktif namun", "pengecekan data"]):
            return "Proses Pencarian dan Konfirmasi Data"
        elif any(keyword in text for keyword in ["sesuai", "sama", "telah terdaftar", "sudah dikoreksi","tidak ada perubahan", "sudah disesuaikan", "sudah sepadan", "masih hidup", "sudah sama", "tidak terdapat anomali", "data masih update", "tidak di temukan perbedaan data", "tidak dikoreksi", "nasabah aktif", "fasilitas aktif", "fasilitas masih aktif", "debitur aktif", "data sudah benar"]):
            return "Data Sudah Sesuai"
        elif any(keyword in text for keyword in ["lunas"]):
            return "Sudah Lunas"
        elif any(keyword in text for keyword in ["susah ditemui", "tidak dapat", "tidak bisa dijumpai", "keberadaan", "keberadaannya", "dihubungi", "tidak ditemukan", "tidak memungkinkan","kabur", "mutasi", "merantau", "pindah", "tidak berada", "nasabah tidak ada", "tidak bisa di temukan", "tidak dapat data", "sulit untuk ditemui", "tidak bisa di temukan", "belum bisa jumpai", "tidak memperoleh data terkini", "tidak bisa ditemui"]):
            return "Nasabah Tidak Ditemukan"
        elif any(keyword in text for keyword in ["akan dikoreksi", "mengkoreksi", "di koreksi", "koreksi", "paling lambat", "mengkinikan"]):
            return "Akan Dikoreksi"
        else:
            return "Tidak Terklasifikasi"
    else:
        return "Tidak Terklasifikasi"  # Jika NaN atau bukan string

# Terapkan fungsi pada kolom 'Keterangan'
BT['Cluster'] = BT['Keterangan'].apply(cluster_alasan)

# Tampilkan hasil
BT
```

```python
# Gabungkan dataset FinalData (berisi "Sudah Koreksi") dengan BT
Final = pd.concat([SudahKoreksi, BT], ignore_index=True)

# Perbarui nilai Cluster jika Status_Koreksi adalah "Selesai"
Final.loc[Final['Status_Koreksi'] == "Selesai", 'Cluster'] = "Sudah Koreksi"

# Tampilkan hasil
Final
```

```python
# Daftar kategori yang masuk dalam "Tidak Koreksi"
tidak_koreksi_list = [
    "Data Sudah Sesuai", "Meninggal", "Hapus Buku",
    "Nasabah Tidak Ditemukan", "Sudah Lunas", "Nasabah Asing"
]

# Fungsi untuk mengklasifikasikan Status Koreksi
def klasifikasi_koreksi(cluster):
    if cluster == "Tidak Terklasifikasi":
        return "Tidak Terklasifikasi"
    elif cluster == "Sudah Koreksi":
        return "Sudah Koreksi"
    elif cluster in tidak_koreksi_list:
        return "Tidak Koreksi"
    else:
        return "Belum Koreksi"

# Terapkan fungsi ke kolom 'Cluster' untuk membuat kolom 'Status Koreksi'
Final['Status Koreksi'] = Final['Cluster'].apply(klasifikasi_koreksi)

# Fungsi untuk mengklasifikasikan Status Progress dengan logika yang diperbarui
def klasifikasi_progress(row):
    # Jika cluster ada di tidak_koreksi_list, maka status koreksi adalah "Tidak Koreksi" dan status progress adalah "Done"
    if row['Cluster'] in tidak_koreksi_list:
        return "Done"

    # Jika cluster adalah "Sudah Koreksi", maka status progress adalah "Done"
    elif row['Cluster'] == "Sudah Koreksi":
        return "Done"

    # Jika cluster adalah "Tidak Terklasifikasi", cek status koreksinya
    elif row['Cluster'] == "Tidak Terklasifikasi":
        if row['Status_Koreksi'] == "Tidak Dikoreksi":
            return "Done"
        elif row['Status_Koreksi'] == "Belum Koreksi":
            return "On Progress"
        else:
            return "On Progress"  # Untuk status koreksi lain selain "Tidak Koreksi" dan "Belum Koreksi"

    # Default jika status lainnya, anggap sebagai "On Progress"
    return "On Progress"

# Terapkan fungsi ke setiap baris untuk menentukan status progress berdasarkan status koreksi dan cluster
Final['Status Progress'] = Final.apply(klasifikasi_progress, axis=1)

# Tampilkan hasil
Final[['LJK','Cluster', 'Status Koreksi', 'Status Progress']]

```

```python
# Mengelompokkan data berdasarkan 'LJK' dan 'Status Progress', lalu menghitung jumlah masing-masing status
result = Final.groupby(['LJK', 'Status Progress']).size().unstack(fill_value=0).reset_index()

# Menggabungkan dengan 'Jumlah Anomali' dari data_anomali (LJK = Nama Pelapor)
result = result.merge(data_anomali[['Nama Pelapor', 'Jumlah Anomali']],
                      left_on='LJK', right_on='Nama Pelapor', how='left')

# Menghapus kolom duplikat 'Nama Pelapor' setelah merge
result.drop(columns=['Nama Pelapor'], inplace=True)

# ----> Convert 'Jumlah Anomali' to numeric, handling errors
result['Jumlah Anomali'] = pd.to_numeric(result['Jumlah Anomali'], errors='coerce').fillna(0).astype(int)

# Menambahkan kolom 'Selisih' (Jumlah Anomali - (Done + On Progress))
result['Selisih'] = result['Jumlah Anomali'] - (result.get('Done', 0) + result.get('On Progress', 0))

# Menambahkan kolom 'Status'
# Jika ada selisih lebih dari 0, pastikan status menjadi 'ON PROGRESS'
result['Status'] = result.apply(lambda row: 'ON PROGRESS' if row['Selisih'] > 0 else ('DONE' if row['On Progress'] == 0 else 'ON PROGRESS'), axis=1)

# Pastikan 'Jumlah Anomali' dan 'Selisih' berupa integer
result['Jumlah Anomali'] = result['Jumlah Anomali'].astype('Int64')  # Menggunakan 'Int64' untuk mendukung nilai NaN
result['Selisih'] = result['Selisih'].astype('Int64')

def generate_keterangan(row):
    keterangan = []

    if row['Status'] == 'ON PROGRESS':
        # Ambil data dari Final untuk LJK yang sama dan masih On Progress
        cluster_details = Final[(Final['LJK'] == row['LJK']) &
                                (Final['Status Progress'] == 'On Progress')]['Cluster'].value_counts().to_dict()

        if cluster_details:
            # Menggabungkan informasi cluster dengan format yang sesuai
            cluster_text = ', '.join([f"{v} DATA {k.upper()}" for k, v in cluster_details.items()])
            keterangan.append(cluster_text)

    # Jika ada selisih jumlah anomali, tambahkan keterangan tambahan
    if pd.notna(row['Selisih']) and row['Selisih'] > 0:
        keterangan.append(f"{row['Selisih']} DATA TIDAK DIINFORMASIKAN OLEH LJK/TIDAK SESUAI FORMAT")

    # Gabungkan semua keterangan
    return ', '.join(keterangan)

# Menambahkan kolom 'Keterangan'
result['Keterangan'] = result.apply(generate_keterangan, axis=1)

# Menampilkan hasil akhir
result
```

VISUALISASI

```python
import matplotlib.pyplot as plt

# Menghitung jumlah setiap kategori dalam "Status"
summary = result['Status'].value_counts()

# Pastikan warna sesuai dengan label
color_mapping = {
    "DONE": "#4CAF50",  # Hijau
    "ON PROGRESS": "#FFC107",  # Kuning
}

# Ambil warna berdasarkan urutan label di summary.index
colors = [color_mapping[label] for label in summary.index]

# Membuat pie chart dengan warna yang sesuai
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    summary,
    labels=summary.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,  # Warna sesuai urutan label
    wedgeprops={'edgecolor': 'black'}
)

# Tambahkan legenda di samping dengan jumlah masing-masing kategori
legend_labels = [f"{label}: {value}" for label, value in zip(summary.index, summary.values)]
ax.legend(wedges, legend_labels, title="Jumlah", loc="center left", bbox_to_anchor=(1, 0.5))

# Tambahkan judul
plt.title('Status Tindak Lanjut LJK')

# Tampilkan chart
plt.show()
```

```python
# Filter data hanya untuk "On Progress"
on_progress_data = Final[Final['Status Progress'] == 'On Progress']

# Hitung jumlah per kategori dalam "Cluster"
on_progress_summary = on_progress_data['Cluster'].value_counts()

# Membuat bar chart
plt.figure(figsize=(10, 6))
plt.bar(on_progress_summary.index, on_progress_summary.values, color='#FFC107')  # Warna kuning untuk On Progress

# Menambahkan label dan judul
plt.xlabel("Kategori", fontsize=12)
plt.ylabel("Jumlah", fontsize=12)
plt.title("Distribusi Keterangan untuk Status On Progress", fontsize=14)
plt.xticks(rotation=45, ha='right')  # Memiringkan label x agar lebih mudah dibaca

# Menampilkan angka di atas bar
for i, value in enumerate(on_progress_summary.values):
    plt.text(i, value + 2, str(value), ha='center', fontsize=10, fontweight='bold')

plt.show()
```
